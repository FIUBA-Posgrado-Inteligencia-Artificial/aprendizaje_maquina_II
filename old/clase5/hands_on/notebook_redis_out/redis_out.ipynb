# Ejemplo de predicción de modelo usando Valkey

Este notebook muestra cómo obtener predicciones de un modelo que produce predicciones en lotes. Las predicciones se cargaron en Valkey. 

En este notebook, recuperamos las predicciones del modelo almacenadas en Valkey. Los datos deben ser leídos, convertidos a cadenas y luego hasheados. Con este enfoque, podemos verificar si los datos existen en Valkey y recuperar la predicción correspondiente. En caso de que los datos no existan, asignamos un valor de cero. Esta estrategia simula cómo podría comportarse un servicio en producción ante casos no contemplados.

La ventaja de utilizar Valkey en este contexto radica en su capacidad para almacenar datos de forma eficiente en memoria, lo que permite un acceso rápido a las predicciones previamente calculadas. Además, Valkey ofrece funcionalidades de almacenamiento en caché y persistencia de datos, lo que garantiza la disponibilidad y la integridad de las predicciones incluso en entornos de producción de alta demanda.

```python
def predict(largo_petalo, ancho_petalo, largo_sepalo, ancho_sepalo):
    attributes = [largo_petalo, ancho_petalo, largo_sepalo, ancho_sepalo]
    substring = " ".join(map(str, [round(float(x),1) for x in attributes]))
    key = hashlib.sha256(substring.encode()).hexdigest()

    output = r.get(key)

    if output:
        return r.get(key)
    #else:
        # output = model.predict(x)
        # queue.insert(x) 

    return output
```

```python
# Obtención de datos para prueba

df_temp = pd.read_csv("./iris.csv", header=None)

# Sampleamos 100 valores al azar
df_temp = df_temp.sample(200)

test_values = df_temp.values.tolist()
# Conversión de valores a cadenas y hash
# Esto debería implementarse en el pipeline. Dado que los números de punto flotante pueden 
# presentar problemas debido a pequeñas variaciones, se podría considerar redondearlos.
string_values = [' '.join(map(str, sublist)) for sublist in test_values]
hashed_values = [hashlib.sha256(substring.encode()).hexdigest() for substring in string_values]  

# Obtención de las salidas del modelo desde Valkey
# Estas salidas ya fueron procesadas en el pasado, por lo que consultamos Valkey para obtenerlas.
model_output = [r.get(hash_key) for hash_key in hashed_values]

# Reemplazo de valores nulos con un valor predeterminado
# Esto es necesario porque, en caso de una nueva entrada que el modelo no haya visto durante 
# el procesamiento por lotes, necesitamos proporcionar una salida. Esta entrada podría ser
# encolada para su posterior procesamiento en lotes.
model_output = [value if value is not None else "setosa" for value in model_output]
```

Veamos la salida del modelo para diferentes entradas:

```python
for index, test_value in enumerate(test_values):
    print(f"Para la entrada: {test_value}")
    
    print(f"El modelo predice que: {model_output[index]}")
```

```python
# Representemos graficamente a estos datos
test_values_df = pd.DataFrame(test_values)
test_values_df = pd.concat([test_values_df, pd.Series(model_output)], axis=1)
test_values_df.columns=["0", "1", "2", "3", "label"]
sns.scatterplot(x="2", y="3", hue="label", data=test_values_df)
plt.title("Model output for test values");
```
