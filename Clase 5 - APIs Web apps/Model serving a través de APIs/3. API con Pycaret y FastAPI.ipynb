# API con Pycaret y FastAPI

#### Obtener predicciones con requests

```python
import requests

def get_predictions(carat_weight, cut, color, clarity, polish, symmetry, report):
    url = 'http://localhost:8000/predict?carat_weight={carat_weight}&cut={cut}&color={color}&clarity={clarity}&polish={polish}&symmetry={symmetry}&report={report}'\
    .format(carat_weight = carat_weight, cut = cut,\
     color = color, clarity = clarity, polish = polish, symmetry = symmetry, report = report)
    
    x = requests.post(url)
    print(x.text)
```

```python
get_predictions(1.1, 'Ideal', 'H', 'SII', 'VG', 'EX', 'GIA')
```

### Creación de API automatizada

```python
# load the dataset from pycaret
from pycaret.datasets import get_data
data = get_data('diamond')
```

```python
# initialize setup
from pycaret.regression import *
s = setup(data, target = 'Price', 
          transform_target = True, 
          log_experiment = True, 
          #log_plots = True,
          experiment_name = 'diamond')
```

```python
# train model
lightgbm = create_model('lightgbm')
```

```python
# create API
create_api(lightgbm, 'my_first_api')

# Run the API
!python my_first_api.py
```
